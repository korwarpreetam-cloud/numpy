### list vs array

```python
py_list=[1,2,3]#list multiplicatioin
print(py_list*2)

np_array= np.array([1,2,3])#element wise multiplicatopn
print(np_array*2)

import time 
start = time.time()
py_list=[i*2 for i in range(100000)]
print(time.time()-start)

start = time.time()
np_array=np.arange(1000000)*2
print(time.time()-start)
```

### creating array from scratch


```python
zeros=np.zeros((3,4))
print(zeros)

ones = np.ones((3,4))
print(ones)

full = np.full((2,2),7)
print(full)

random = np.random.random((2,3))
print(random)

sequence=np.arange(0,10,1)
print(sequence)
```

### vector matrix and tensor


```python
vector = np.array([1,2,3])
print(vector)

matrix = np.matrix([[1,2,3],[1,2,3]])
print(matrix)

#tensor -- multiple dimensins
```

### array prop

```python
arr= np.array([[1,2,3],[1,22,3]])
print(arr.shape)
print(arr.ndim)
print(arr.size)
print(arr.dtype)
```

### array reshaping

```python
arr=np.arange(12)
print(arr)

reshapes = arr.reshape((3,4))
print(reshapes)

raveled = reshapes.ravel()#return view instead of copy
print(raveled)

#transpose
transpose=reshapes.T
print(transpose)
```
