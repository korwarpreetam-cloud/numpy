```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# data structure [resturant_id, 2021, 2022, 2023, 2024]
sales_data=np.array([
    [1,150000,180000,220000,250000], 
    [2,120000,140000,160000,190000], 
    [3,200000,230000,260000,300000],
    [4,180000,210000,240000,270000],
    [5,160000,185000,205000,230000],

]) 

print("===zomato analysis===")
print("sales data shape", sales_data.shape)
print("sample data of three resto",sales_data[0:3])
print(sales_data[:, 1:])
```

```python
# total sales per year
print("year wise sales", np.sum(sales_data, axis=0))
yearly_sales= np.sum(sales_data[:,1:],axis=0)
print(yearly_sales)
print(sum(yearly_sales))



# minimum sales per restaruent
min_sales=np.min(sales_data[:, 1:], axis=1)
print(min_sales)

# minimum sales per year
max_sales=np.max(sales_data[:, 1:], axis=0)
print(max_sales)

# average sles of restaruen 
avg_sales = np.mean(sales_data[:,1:],axis=1)
print(avg_sales)
```

```python
cumsum = np.cumsum(sales_data[:,1:],axis=1)
print(cumsum) 
plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("average cumulative sales per year across all restarount ")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```
